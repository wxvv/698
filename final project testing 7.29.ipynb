```python
from breezypythongui import EasyFrame
import math 

class hedron(EasyFrame):
    color = '#A4D3EE'
    def __init__(self):
        EasyFrame.__init__(self, title = "Rhombicosidodecahedron Calculator", 
                           width = 520, height = 410)
        #self["background"] = "#EDEDED"
        
        self.addLabel(text = "Do you know what a rhombicosidodecahedron is?", 
                      row = 0, column = 0)  
        # button 1
        aaa = self.addButton(text = 'Yes', row = 1, column = 0, command = self.yes)
        aaa.config(fg = '#030303')
      
        # button 2
        bbb = self.addButton(text = 'No', row = 1, column = 1, columnspan = 2, command = self.no)
        bbb.config(fg = '#7F7F7F')
       
        
             
        self.addLabel(text = 'Edge Length', row = 2, column = 0, background = hedron.color)
        self.edgeField = self.addFloatField(value = 0.0, row = 2, column = 1, width=12)
        
        self.addLabel(text = 'Output :', row = 3, column = 0)
        
        self.addLabel(text = 'Surface Area \n(≈59.3059828*edge^2)', row = 4, column = 0, background = hedron.color)
        self.areaField = self.addFloatField(value = 0.0, row = 4, column = 1, precision = 3)
        
        self.addLabel(text = 'Volume \n(≈41.6153238*edge^3)', row = 5, column = 0, background = hedron.color)
        self.volumeField = self.addFloatField(value = 0.0, row = 5, column = 1, precision = 3)
        
        kkk = self.addButton(text = 'Compute', row = 6, column = 0, columnspan = 2, command = self.computeAV)
        kkk.config(fg='#050505')
        
            
    def yes(self):
        self.messageBox(title = "Good", 
                        message = "You are so smart! \nLet's do some calculation then.", width = 60)
    
    def no(self):
        self.messageBox(title = "Story time", 
                        message = "In geometry, the rhombicosidodecahedron, is an Archimedean solid, one of thirteen convex isogonal nonprismatic solids constructed of two or more types of regular polygon faces.\nIt has 20 regular triangular faces, 30 square faces, 12 regular pentagonal faces, 60 vertices, and 120 edges. \nAlright let's do some calculation then.",
                       width = 100)
        
    def computeAV(self):
        try:
            edge = self.edgeField.getNumber()
            if edge <= 0: 
                self.messageBox(title = "Error", message = "Edge must be > 0.")
            else:
                area = edge**2 * ( 30 + 5*math.sqrt(3) + 3*math.sqrt(25 + 10*math.sqrt(5) ))
                volume = 1 / 3 * (edge**3 * ( 60 + 29 * math.sqrt(5)))
                self.areaField.setNumber(area)
                self.volumeField.setNumber(volume)
                
        except ValueError:
            self.messageBox(title = "ERROR", message = "That's not an EDGE!")

            
hedron().mainloop()
```
